```python
import pandas as pd
import numpy as np

rng = np.random.RandomState(0)
indices = rng.choice(np.arange(housing.shape[0]), size=500,
                     replace=False)

columns_drop = ["longitude", "latitude"]
subset = housing.iloc[indices].drop(columns=columns_drop)

subset["valor_media_casa"] = pd.qcut(subset["valor_media_casa"], 6, retbins=False)
subset["valor_media_casa"] = subset["valor_media_casa"].apply(lambda x: x.mid)
```

```python
_ = sns.pairplot(data=subset, hue="valor_media_casa", palette="rocket")
```

One hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown = 'ignore', sparse = False)
df_ohe = pd.DataFrame(ohe.fit_transform(housing[['proximidade_oceano']]))
```

```python
df_ohe.index = housing.index
df_num = housing.drop(['proximidade_oceano'], axis=1)
df2 = pd.concat([df_num, df_ohe], axis=1)
df2.head()
```

```python
y = df2.valor_media_casa
X = df2.drop(columns = 'valor_media_casa')
```

```python
X.head()
```

ML

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

regressão linear

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_test.mean()
```

```python
y_pred.mean()
```

MSE Erro quadrático médio


```python
from sklearn.metrics import mean_squared_error

housing_predictions = model.predict(x_train)
lin_mse = mean_squared_error(y_train, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

MAE Erro Absoluto Médio

```python
from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(y_train, housing_predictions)
lin_mae
```

arvore de decisão

```python
from sklearn.tree import DecisionTreeRegressor

model2 = DecisionTreeRegressor(random_state=42)
model2.fit(x_train, y_train)
```

```python
housing_predictions2 = model2.predict(x_train)
tree_mse = mean_squared_error(y_train, housing_predictions2)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
from sklearn.svm import SVR

model3 = SVR(kernel="linear")
model3.fit(x_train, y_train)
housing_predictions3 = model3.predict(x_train)
svm_mse = mean_squared_error(y_train, housing_predictions3)
svm_rmse = np.sqrt(svm_mse)
svm_rmse
```
